```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Case studies
map_res = ['1.8', '3', '5']
```

```python
# High res
filepath = './simple_mdff_vds_8_replicas_'+ map_res[0] + 'angstrom_20iter_5k_numsteps/'
common_str = '.Calculating_the_cross-correlation_coefficient/cc.dat'
master = []
for rep in range (0,8):
    replica = 'simple-mdff-replica-'+ str(rep) + '/'
    for iter in range(0,10):
        current_file = filepath + replica + str(iter) + common_str
        #print(current_file)
        if iter == 0:
            df = pd.read_csv(current_file,delimiter=' ',header=None)
            df['iteration'] = iter
            df['replica'] = rep
        else:
            df = pd.read_csv(current_file,delimiter=' ',header=None)
            df['iteration'] = iter
            df['replica'] = rep
#             df = df.append(dftemp)
        master.append(df)
df = pd.concat(master,axis=0)
            
df.columns = ['First CCC', 'Last CCC', 'Iteration_index', 'Replica_index']

#Plotting
plt.figure(num=None, figsize=(8, 6), dpi=300, facecolor='w', edgecolor='k')
for repindex, df_repindex in df.groupby('Replica_index'):
    plt.plot(df_repindex['Iteration_index'], df_repindex['Last CCC'],linewidth=1, label=repindex)
    plt.xticks(np.arange(min(df_repindex['Iteration_index']), max(df_repindex['Iteration_index'])+1, 1.0))
    plt.legend()
plt.xlabel('Iteration index', fontsize=16)
plt.ylabel('Last frame CCC', fontsize=16)
plt.title('Case:' + map_res[0] + ' $\AA$ - 8 replicas - 20 iterations per replica - 5000 steps/iteration', fontsize=12)
plt.savefig(map_res[0] + 'A-8replicas-20iters.png',dpi=300)
```
